# Pandas vs Spark 

# 1- Spark:

```python
from pyspark.sql import SparkSession
import pyspark.sql.functions as F

SEED = 42

N = 100_000_000
CITIES = ["Khartoum", "Madani", "AL-Jazeera", "Dongola", "Kasala"]

spark = SparkSession.builder \
    .appName("PandasVsSpark") \
    .master("local[*]") \
    .config("spark.driver.memory", "10g") \
    .getOrCreate()

city_array = F.array(*[F.lit(c) for c in CITIES])

df_spark = (
    spark.range(N)
    .withColumn(
        "city",
        F.element_at(city_array, (F.rand(seed=SEED) * len(CITIES)).cast("int") + 1)
    )
    .withColumn("sales", F.rand(seed=SEED) * 1000)
    .withColumn("units", (F.rand(seed=SEED) * 9 + 1).cast("int"))
)

_ = df_spark.count()

spark.stop()
```

# 2- Pandas :

```python
import pandas as pd
import numpy as np

np.random.seed(42)


df_pd = pd.DataFrame({
    "city": np.random.choice(CITIES, size=N),
    "sales": np.random.rand(N) * 1000,
    "units": np.random.randint(1, 10, size=N)
})
```
